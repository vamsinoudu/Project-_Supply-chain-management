# Conducting Hypothesis Testing on supply chain management dataset

```python
import numpy as np
import pandas as pd
```

# workers_num

```python
#Chi2_Square test:
from scipy.stats import chi2_contingency
hypothesis=pd.crosstab(df['workers_num'],df['product_wg_ton'])
chi2_stat,p,dof,expected=chi2_contingency(hypothesis)
print(f"chi2_value:   {chi2_stat:0.5g}")
print(f"p_value: {p:0.5g}")
print(f"dof_value:  {dof:0.5g}")
```

```python
#two sample Z test:
from statsmodels.stats.weightstats import ztest as ztest
hyp=pd.crosstab(df['workers_num'],df['product_wg_ton'])
z,pval=ztest(hyp)
print("z_value: ",z)
print("p_Value: ",p)
```

```python
if pval.all()> 0.05:
    print("It fail to Reject Null Hypothesis")
else:
    print("It Reject Null Hypothesis")
```

# Conducting Hypothesis Testing
# Null Hypothesis = Workers_num is impacting on target variable 
# Alternative Hypothesis = Workers_num is NOT impacting on target variable
# Here P>0.05 so 
Null Hypothesis Accepted and Alternative Hypothesis Rejected due to P value not less than 0.005
